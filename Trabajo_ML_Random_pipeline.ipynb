# Proyecto para predecir la quiebra de una empresa




ML_prevision_de_quiebra

El objetivo de este proyecto es predecir si una empresa está en riesgo de quiebra, utilizando datos financieros históricos. El objetivo es identificar los factores más influyentes y construir un modelo predictivo.

Dataset
Nombre: Polish Companies Bankruptcy Data
Tamaño: 6819 observaciones y 96 columnas
Variable objetivo: Bankrupt? (1 = empresa quebrada, 0 = no quebrada)

Cada fila representa una empresa y contiene 95 ratios financieros anónimos.

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn import linear_model, metrics, model_selection
from sklearn.linear_model import ElasticNet
import matplotlib.pyplot as plt
from sklearn.linear_model import Ridge
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.metrics import accuracy_score, f1_score, precision_score, recall_score
from sklearn.metrics import roc_curve, roc_auc_score
from sklearn.metrics import precision_recall_curve, average_precision_score
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.pipeline import Pipeline
from xgboost import XGBClassifier
```

```python
df = pd.read_csv('data.csv')

df.info()
df.describe()
```

```python

df.head()
```

```python
# 2. Preparar datos
X = df.drop('Bankrupt?', axis=1)
y = df['Bankrupt?']

# 3. Escalar las variables (recomendado para SVM)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# 4. Dividir datos en entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, stratify=y, test_size=0.3, random_state=42)

```

```python
# 5. Crear pipeline
pipeline = Pipeline([
    ('scaler', StandardScaler()),
    ('xgb', XGBClassifier(use_label_encoder=False, eval_metric='logloss', random_state=42))
])

# 6. Entrenar modelo
pipeline.fit(X_train, y_train)

# 7. Predicción y evaluación
y_pred = pipeline.predict(X_test)
y_proba = pipeline.predict_proba(X_test)[:,1]

print("Matriz de Confusión:\n", confusion_matrix(y_test, y_pred))
print("\nReporte de Clasificación:\n", classification_report(y_test, y_pred))
print("ROC AUC:", roc_auc_score(y_test, y_proba))

# 6. Importancia de variables (del modelo dentro del pipeline)
model = pipeline.named_steps['xgb']
importances = model.feature_importances_
features = X.columns

feat_imp = pd.Series(importances, index=features).sort_values(ascending=False)

# Mostrar top 20 variables
top20 = feat_imp.head(20)
plt.figure(figsize=(10,6))
sns.barplot(x=top20.values, y=top20.index)
plt.title("Top 20 variables más importantes (XGBoost Pipeline)")
plt.xlabel("Importancia")
plt.ylabel("Variable")
plt.tight_layout()
plt.show()
```

```python
# 3. Definir grid de hiperparámetros para buscar
param_grid = {
    'xgb__n_estimators': [50, 100, 200],
    'xgb__max_depth': [3, 5, 7],
    'xgb__learning_rate': [0.01, 0.1, 0.2],
    'xgb__subsample': [0.7, 1],
    'xgb__colsample_bytree': [0.7, 1]
}

# 4. Configurar GridSearchCV
grid_search = GridSearchCV(
    estimator=pipeline,
    param_grid=param_grid,
    cv=5,
    scoring='roc_auc',
    n_jobs=-1,
    verbose=1
)

# 5. Entrenar con búsqueda de hiperparámetros
grid_search.fit(X_train, y_train)

print(f"Mejores hiperparámetros: {grid_search.best_params_}")
print(f"Mejor AUC en CV: {grid_search.best_score_:.4f}")

# 6. Evaluar en conjunto test
y_pred = grid_search.predict(X_test)
y_proba = grid_search.predict_proba(X_test)[:,1]

print("Matriz de Confusión:\n", confusion_matrix(y_test, y_pred))
print("\nReporte de Clasificación:\n", classification_report(y_test, y_pred))
print("ROC AUC en test:", roc_auc_score(y_test, y_proba))

# 7. Importancia variables del mejor modelo
best_model = grid_search.best_estimator_.named_steps['xgb']
importances = best_model.feature_importances_
features = X.columns

feat_imp = pd.Series(importances, index=features).sort_values(ascending=False)
top20 = feat_imp.head(20)

plt.figure(figsize=(10, 6))
sns.barplot(x=top20.values, y=top20.index)
plt.title('Top 20 Variables Más Importantes (XGBoost con GridSearchCV)')
plt.xlabel('Importancia')
plt.ylabel('Variable')
plt.tight_layout()
plt.show()
```

```python
y_pred = model.predict(X_test)

print("Matriz de confusión:")
print(confusion_matrix(y_test, y_pred))

print("\nReporte de clasificación:")
print(classification_report(y_test, y_pred))

print("\nPrecisión global:")
print(accuracy_score(y_test, y_pred))
```
